```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
import time
entry = 4
data = []
x = []
```

```python
for i in range(int(entry)):
    x.append(i+1)
```

```python
prob = []
cases = []
total = 0
minCase  = 1000
maxCase = 0
perm = list(itertools.permutations(x,len(x)))
npermut = len(perm)
```

```python
def isortSteps(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            steps = steps + 3
        steps = steps + 1
        v[j+1] = x
        steps = steps + 4
    steps = steps + 1
    return steps
```

```python
for i in range(0,len(perm)):
        k = isortSteps(perm[i])
        total+=k
        data.append(k)
        cases.append(k)
```

```python
plt.plot(data)
plt.xlabel('Permutation N')
plt.ylabel('Steps')
avg = np.mean(data)
print  ("Average Steps: " + str(avg))
```

```python
plt.hist(cases,20)
plt.xlabel('steps')
plt.ylabel('Repetitions')
plt.legend(['permutations insertion sort '])
```
